```python
import os
import numpy as np 
import pandas as pd 
```

```python
train = pd.read_csv('../data/train.csv')
target = train.target
```

```python
sub_path = "../predictions/train/"
all_files = [x for x in os.listdir(sub_path) if x.endswith('.csv') and 'stacking' not in x]

# Read and concatenate submissions
outs = [pd.read_csv(os.path.join(sub_path, f)).target for f in all_files]
concat_sub = pd.concat(outs, axis=1)
cols = list(map(lambda x: "target_" + str(x), range(len(concat_sub.columns))))
concat_sub.columns = cols
concat_sub.reset_index(inplace=True, drop=True)
concat_sub.head()
```

```python
concat_sub.corr()
```

```python
target_cols = concat_sub.columns
```

```python
# get the data fields ready for stacking
concat_sub['target_max'] = concat_sub.max(axis=1)
concat_sub['target_min'] = concat_sub.min(axis=1)
concat_sub['target_mean'] = concat_sub.mean(axis=1)
concat_sub['target_median'] = concat_sub.median(axis=1)
```

## Mean stacking 

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(target, concat_sub['target_mean'])
```

## Median stacking 

```python
roc_auc_score(target, concat_sub['target_median'])
```

## PushOut + Median Stacking 

```python
cutoff_lo = 0.95
cutoff_hi = 0.05

target_pom = np.where(np.all(concat_sub[target_cols] > cutoff_lo, axis=1), 1, 
                                    np.where(np.all(concat_sub[target_cols] < cutoff_hi, axis=1),
                                             0, concat_sub['target_median']))
```

```python
roc_auc_score(target, target_pom)
```

## MinMax + Mean Stacking 

```python
cutoff_lo = 0.95
cutoff_hi = 0.05

target_mmm = np.where(np.all(concat_sub[target_cols] > cutoff_lo, axis=1), 
                                    concat_sub['target_max'], 
                                    np.where(np.all(concat_sub[target_cols] < cutoff_hi, axis=1),
                                             concat_sub['target_min'], 
                                             concat_sub['target_mean']))
```

```python
roc_auc_score(target, target_mmm)
```

## MinMax + Median Stacking 

```python
cutoff_lo = 0.95
cutoff_hi = 0.05

target_mmmed = np.where(np.all(concat_sub[target_cols] > cutoff_lo, axis=1), 
                                    concat_sub['target_max'], 
                                    np.where(np.all(concat_sub[target_cols] < cutoff_hi, axis=1),
                                             concat_sub['target_min'], 
                                             concat_sub['target_median']))
```

```python
roc_auc_score(target, target_mmmed)
```

## MinMax + BestBase Stacking

```python
target_base = pd.read_csv('../predictions/train/9211_catboost_unique_with_10x_upsample_3103172847.csv').target

cutoff_lo = 0.95
cutoff_hi = 0.05

target_mmb = np.where(np.all(concat_sub[target_cols] > cutoff_lo, axis=1), 
                                    concat_sub['target_max'], 
                                    np.where(np.all(concat_sub[target_cols] < cutoff_hi, axis=1),
                                             concat_sub['target_min'], 
                                             target_base))
```

```python
roc_auc_score(target, target_mmb)
```
